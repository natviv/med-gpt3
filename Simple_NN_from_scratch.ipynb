```
import numpy as np
W = np.eye(3)
x = np.arange(3)

print(f"W shape {W.shape}")
print(f"x shape {x.shape}")
print(f"x {x}")
print(f"W.x {np.dot(W, x)}")


def sigmoid(W, x):
  return 1/(1 + np.exp(-np.dot(W, x)))


z = sigmoid(W, x)
print(f"z {z}")
```

```
import matplotlib.pyplot as plt
import math
  
x = np.linspace(-10, 10, 100)
z = 1/(1 + np.exp(-x))
  
plt.plot(x, z)
plt.xlabel("x")
plt.ylabel("sigmoid(x)")
  
plt.show()
```

```
dz = z*(1-z)

plt.plot(x, dz)
plt.xlabel("x")
plt.ylabel("sigmoid'(x)")
  
plt.show()
```
